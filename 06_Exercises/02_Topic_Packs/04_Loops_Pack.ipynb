## Lo3: Count Evens
**Task:** Count how many even numbers are in the list and print the count.

```python
nums = [1,2,3,4]
# expected: 2
```


### Hints
- Use for/while loops.
- Use break/continue when needed.
- Print outputs as shown.


````python
nums = [1,2,3,4]

count = 0

for num in nums:
    if not num % 2:
        print(num)
        count += 1
print(count)
````

### Solution


````python
nums = [1,2,3,4]
count = 0
for n in nums:
    if n % 2 == 0:
        count += 1
print(count)
````
